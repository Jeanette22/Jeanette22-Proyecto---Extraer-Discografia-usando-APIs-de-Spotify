### Uso del ID obtenido para obtener toda ld discografia del artista

```python
# Recordar que el token en determinado momento experiraba, por eso 
# hacemos una funcion para que se renueve 
```

```python
import base64 #client id y secret nos permite codificar para acceder a la api de autorizacion de spotify
```

```python
def get_token(client_id, client_secret):
    encoded = base64.b64encode(bytes(client_id+':'+client_secret, 'utf-8'))
    params = {'grant_type':'client_credentials'}
    header={'Authorization': 'Basic ' + str(encoded, 'utf-8')}
    r = requests.post('https://accounts.spotify.com/api/token', headers=header, data=params)
    if r.status_code != 200:
        print('Error en la request.', r.json())
        return None
    print('Token válido por {} segundos.'.format(r.json()['expires_in']))
    return r.json()['access_token']
#errores que tuve: espacio en el 'Basic ', 
#en encoded = base64.base64encode habia puesto decode
#en return r.json habia puesto access token, y es access_token, GUION BAJO!
```

```python
# def obtener_discografia(artist_id, token, return_name = False, page_limit = 50, country= None): #recordar los : al final
#     url = f'https://api.spotify.com/v1/artists/{artist_id}/albums' #Nunca dejar espacio en f'https (recordar)
#     header = {'authorization': f'Bearer {token}'}
#     params = {'limit': page_limit, 
#               'offset': 0,
#               'country': country}

#     lista = []
#     r = requests.get (url, params = params, headers = header)
#     if r.status_code != 200:
#         print('Error en request.', r.json())
#         return None
    
    
#     if return_name: #guion bajo 
#         lista += [(item['id'], item['name']) for item in r.json()['items']]       #le agregamos una lista por comprension 
#     else: 
#         lista += [item['id'] for item in r .json()['items']]
    
                       
#     while r.json()['next']:
#         r= requests.get(r.json()['next'], headers = header) #Tener en cuenta los parentesis, ante errores 
#         if r.status_code != 200:
#             print ("Error en la request", r.json())
#             return None 
                        
#         if return_name:
#             lista += [(item['id'], item ['name']) for item in r.json()['items']]       #le agregamos una lista por comprension 
#         else: 
#             lista += [item ['id'] for item in r.json()['items']]
                       
#     return lista

def obtener_discografia(artist_id, token, return_name=False, page_limit=50, country=None):
    url = f'https://api.spotify.com/v1/artists/{artist_id}/albums'
    header = {'Authorization': f'Bearer {token}'}
    params = {'limit': page_limit, 
              'offset': 0,
              'country': country}
    lista = []
    r = requests.get(url, params=params, headers=header)
    
    if r.status_code != 200:
        print('Error en request.', r.json())
        return None
    
    if return_name:
        lista += [(item['id'], item['name']) for item in r.json()['items']]
    else:
        lista += [item['id'] for item in r.json()['items']]
        
    while r.json()['next']:
        r = requests.get(r.json()['next'], headers=header) # El resto de los parámetros están dentro de la URL
        if return_name:
            lista += [(item['id'], item['name']) for item in r.json()['items']]
        else:
            lista += [item['id'] for item in r.json()['items']]
    
    return lista
```

```python
# """Función que devuelve una lista con los tracks de un album determinado"""
# def obtener_tracks (album_id, token, return_name= False, page_limit = 50, market= None): ##arriba
#     url = f'https://api.spotify.com/v1/albums/{album_id}/tracks'         #CAMBIAMOS EL URL A COMPARACIÓN CON LA ANTERIOR DE ALBUMS
#     header = {'Authorization': f'Bearer{token}'}
#     params = {'limit': page_limit, 
#               'offset': 0,
#               'market': market} #en album es coutry, aca usamos market
  
    
    
#     lista = []
#     r = requests.get (url, params = params, headers = header)
#     ###########
#     if r.status_code != 200:
#         print('Error en request.', r.json())
#         return None
#     ####################
#     if return_name: #guion bajo 
#         lista += [(item['id'], item['name']) for item in r.json()['items']]       #le agregamos una lista por comprension 
#     else: 
#         lista += [item['id'] for item in r.json()['items']]

        
#     while r.json()['next']:
#         r= requests.get(r.json()['next'], headers = header) #Tener en cuenta los parentesis, ante errores 
#         if return_name:
#             lista += [(item['id'], item['name']) for item in r.json()['items']]       #le agregamos una lista por comprension 
#         else: 
#             lista += [item['id'] for item in r.json()['items']]    #Mensaje de error de status, abajo el rjson por eso el \n,                
#     return lista                                         #reduce un paso y codigo por el if r.status_code usado antes 
    
def obtener_tracks(album_id, token, return_name=False, page_limit=50, market=None):
    url=f'https://api.spotify.com/v1/albums/{album_id}/tracks'
    header = {'Authorization': f'Bearer {token}'}
    params = {'limit': page_limit, 
              'offset': 0,
              'market': market}
    lista = []
    r = requests.get(url, params=params, headers=header)
    
    if r.status_code != 200:
        print('Error en request.', r.json())
        return None
    
    if return_name:
        lista += [(item['id'], item['name']) for item in r.json()['items']]
    else:
        lista += [item['id'] for item in r.json()['items']]
        
    while r.json()['next']:
        r = requests.get(r.json()['next'], headers=header) # El resto de los parámetros están dentro de la URL
        if return_name:
            lista += [(item['id'], item['name']) for item in r.json()['items']]
        else:
            lista += [item['id'] for item in r.json()['items']]
    
    return lista
```
